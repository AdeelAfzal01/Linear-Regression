```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import linear_model
```

```python
df = pd.DataFrame({
    'area' : [2600, 3000, 3200, 3600, 4000],
    'price' : [550000, 565000, 610000, 680000, 725000]
})
```

```python
%matplotlib inline
plt.xlabel('Area (sqr ft)')
plt.ylabel('Price ($)')
plt.scatter(df.area, df.price, color='red', marker='+')
```

```python
# create an object for linear regression
reg = linear_model.LinearRegression()
reg.fit(df[['area']], df.price)
```

```python
reg.predict([[3300]])
```

```python
# y(price) = m(coeficient=slope) * x(area) + b (Y intercept)
# linear model calculates the values m and b

print (reg.coef_)
print (reg.intercept_)
```

```python
# predict a 2D array
d = pd.DataFrame({'areas' : [1000, 1500, 2300, 3540, 4120, 4560, 5490, 3460, 4750, 2300, 9000, 8600, 7100]})
p = reg.predict(d)
```

```python
d['prices'] = p
d.to_csv('predictions.csv', index=False)
```

```python
%matplotlib inline
plt.xlabel('area', fontsize=20)
plt.ylabel('price', fontsize=20)
plt.scatter(df.area, df.price, color='red', marker='+')
plt.plot(df.area, reg.predict(df[['area']]), color='blue')
```

# predict the per capita income

```python
per_capita_income = pd.DataFrame({
    'year' : ['1970','1971','1972','1973','1974','1975','1976','1977','1978','1979','1981','1982','1983','1984','1985',
             '1986','1987','1988','1989','1990','1991','1992','1993','1994','1995','1996','1997','1998','1999','2000',
             '2001','2002','2003','2004','2005','2006','2007','2008','2009','2010','2011','2012','2013','2014','2015',
             '2016','2017','2018','2019',],
    'income' : [3399,3768,4251,4804,5576,5998,7062,7100,7247,7602,8335,9434,9619,10416,10790,11018,
                                11482,12974,15080,16426,16838,17266,16412,17500,17750,18100,18600,18900,19321,19677,
                                19990,20188,20800,21198,21800,22900,23789,23999,24190,24990,25900,26800,26989,
                                27900,28700,29500,31000,32000,33000]
})
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('Per Capita Income (US$)')
plt.scatter(per_capita_income.year,per_capita_income.income, color='red', marker='+')
```

```python
income_model = linear_model.LinearRegression()
income_model.fit(per_capita_income[['year']],per_capita_income.income)
```

```python
income_model.predict([[2020]])
```

### Save the model using pickle

# Save the model

```python
import pickle
```

```python
with open('model_pickle','wb') as f:
    pickle.dump(income_model,f)
```

# Load the saved model

```python
with open('model_pickle','rb') as f:
    load_model = pickle.load(f)
```

```python
load_model.predict([[2020]])
```

### Save the model using joblib

```python
from sklearn.externals import joblib
```

```python
joblib.dump(income_model,'model_joblib')
```

```python
jm = joblib.load('model_joblib')
```

```python
jm.predict([[2020]])
```
